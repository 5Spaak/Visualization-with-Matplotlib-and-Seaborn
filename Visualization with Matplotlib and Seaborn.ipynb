# Visualization with Matplotlib and Seaborn

````python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
````

## Line Chart

````python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
````

````python
plt.figure(figsize=(13,6))
plt.plot(yield_apples);
````

## Customizing the X-axis

````python
years = [2010,2011,2012,2013,2014,2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
````

````python
plt.figure(figsize=(13,6))
plt.plot(years,yield_apples)
````

## Axis Labels

````python
plt.figure(figsize=(13,6))

plt.plot(years,yield_apples)

plt.xlabel('Year')
plt.ylabel('Yield (ton per hectare)');
````

````python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
````

````python
sns.set_style("whitegrid")

plt.figure(figsize=(13,6))

plt.plot(years, apples, color="black", linestyle="solid", linewidth=5)
plt.plot(years, oranges, color='purple', linestyle="dashed", linewidth=5)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
plt.title("Cross Yields in Kanto)")
plt.legend(['Apples','Oranges'])
````

````python
sns.set_style("whitegrid")

plt.figure(figsize=(13,6))

plt.plot(years, apples, color='blue', linestyle='solid', linewidth=4)
plt.plot(years, oranges, color='orange', linestyle='dashed', linewidth=4)
plt.xlabel("Years")
plt.ylabel("Yield (tons per hectare)")
plt.legend(["Apples","Orange"]);
plt.title("Crop Yields in Kanto");
````

````python
sns.set_style("whitegrid")

plt.figure(figsize=(13,6))

plt.plot(years, apples, marker="o", linestyle='solid', linewidth=3, markersize=5)
plt.plot(years, oranges, marker="x", ls='dashed', lw=5, ms=7,mec='black', mew=7, mfc="pink")
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title("Crop Yields in Kanto")
plt.legend(["Apples","Oranges"]);
````

````python
plt.figure(figsize=(13,6))

plt.plot(years, apples, marker="s", c="b", ls="-", lw=3, ms=9, mec="navy")
plt.plot(years, oranges, marker="o", c="r",ls="--", lw=3, ms=10, alpha=0.6)

plt.xlabel("Years")
plt.ylabel("Yield (tons per hectare)")

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
````

The `fmt` argument provides a shorthand for specifying the marker shape, line style, and line color. It can be provided as the third argument to `plt.plot`.

```
fmt = '[marker][line][color]'
```


````python
plt.figure(figsize=(13,6))

plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel("Years")
plt.ylabel("Yield (tons per hectare)")
plt.legend(["Apples", "Oranges"])
````

````python
plt.figure(figsize=(13,6))

plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare) ");
````

### Changing the Figure Size 

````python
plt.figure(figsize=(13,6))
plt.figure(figsize=(12,6))
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)")
````

````python
plt.figure(figsize=(13,6))

plt.plot(years, apples, 's-b',lw=3, ms=12,mec='black', mfc='navy')
plt.plot(years, oranges, 'o--r' ,lw=3, ms=12, alpha=0.7)

plt.xlabel("Years")
plt.title("Yield of Oranges (tons per hectare) ");
plt.ylabel('Yield (tons per hectare)');
plt.legend(['Apples','Oranges']);
````
